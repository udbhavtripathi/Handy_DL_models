```python
A = 1
B = 1
C = 1


ahat = A*x**2 + B*x + C
```

```python
import numpy as np


# Initial parameters
A = 1.0
B = 1.0
C = 1.0

# Number of iterations
iterations = 100

# Learning rate
learning_rate = 0.001

for i in range(iterations):
    ahat = A * x**2 + B * x + C

    # Compute the loss
    loss = np.mean((ahat - a)**2)  

    print("_______LOSS________", loss)

    # Compute the gradients
    dl_d_A = 2 * np.mean((ahat - a) * x**2) 
    dl_d_B = 2 * np.mean((ahat - a) * x) 
    dl_d_C = 2 * np.mean(ahat - a)

    # Update the parameters using gradient descent
    A = A - learning_rate * dl_d_A
    B = B - learning_rate * dl_d_B
    C = C - learning_rate * dl_d_C

    print(f"Iteration {i + 1}: Loss = {loss}, A = {A}, B = {B}, C = {C}")
```
